```python
Fs, data1 = wavfile.read('test.wav')
data=[]
for i in range(1 , len(data1)):
   v = (data1[i][0] + data1[i][1]) / 2
   data.append(v)

data = np.abs(data)
x = np.linspace(0, len(data)/Fs , len(data))
# plt.plot(data)

plt.plot(data)
```

```python
n=200 

An=[] 

Bn=[]

sum=0
print(len(data))
for i in range(n):

    an=np.sum(data*np.cos(i*x*2*np.pi/(len(data)/Fs)))*(1.0/(len(data/(Fs+0.00001))))

    An.append(an)

for i in range(n):

     bn=np.sum(data*np.sin(i*x*2*np.pi/(len(data)/Fs)))*(1.0/(len(data/(Fs+0.000001))))

     Bn.append(bn) 

Cn = []
for k in range(len(Bn)):
      Cn.append(np.sqrt(An[k]**2 + Bn[k]**2))

plt.plot(Cn)


plt.show()
```

and here we make our signal:

```python
for i in range(n):

     if i==0.0:

         sum=sum+An[i]/2

     else:

         sum=sum+(An[i]*np.cos(i*x)+Bn[i]*np.sin(i*x))

plt.plot(x,sum,'g')



plt.title("fourier series for square wave")
```
